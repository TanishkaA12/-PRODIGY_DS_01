```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
file_path="/content/API_SP.POP.TOTL_DS2_en_csv_v2_1176870.csv"
data=pd.read_csv(file_path, skiprows=4)
```

```python
print("Dataset Preview:")
print(data.head())
```

```python
#select year (ex:2020)
year="2020"
data_year= data[["Country Name",year]].dropna()
data_year = data_year.rename(columns={year:"Population"})
```

```python
#--------Bar Chart: Top 10 most populated countries--------
top10= data_year.sort_values(by="Population",ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x="Population",y="Country Name", data=top10, palette="mako")
plt.xlabel(f"Top 10 Countries by Population in {year}")
plt.xlabel("Population")
plt.ylabel("Country")
plt.show()
```

```python
# ----- Histogram: Distribution of population across all countries -----
plt.figure(figsize=(8,5))
sns.histplot(data_year["Population"], bins=30, kde=True, color="skyblue")
plt.title(f"Population Distribution Across Countries in {year}")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```
